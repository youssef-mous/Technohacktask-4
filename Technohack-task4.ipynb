# Pivote Table

```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
```

```python
dataset = pd.read_csv('iris.csv')
```

```python
dataset.head(15)
```

```python
#average of SepalLength for Irsi-setosa species 
dataset[dataset['Species']=='Iris-setosa'].SepalLengthCm.mean()
```

```python
#average of SepalLength for Iris-versicolor  species 
dataset[dataset['Species']=='Iris-versicolor'].SepalLengthCm.mean()
```

```python
# groupby function may help but it's limited 
dataset.groupby('Species')['SepalLengthCm'].mean()
```

```python
#calculate all the data mean, std, min, max, median, Count and sum  of all the data for each species
# now this is a little hard , so we must use pivot_table
dataset.pivot_table(index='Species' ,values=['SepalLengthCm','SepalWidthCm','PetalLengthCm','PetalWidthCm'],aggfunc=['mean','std','min','max','median','count','sum'])
```
